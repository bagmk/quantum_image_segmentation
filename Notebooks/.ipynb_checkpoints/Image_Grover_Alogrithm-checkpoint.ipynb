```python
import time
n_samples_show = 49
fig, axes = plt.subplots(nrows=15, ncols=8, figsize=(20, 20))
x = range(n_samples_show)
y = []
for i in range(n_samples_show):
    imgdata=imageResize(callImage(i+1,path,name),32)
#     imgdata=imageBinarize(imgdata, 110)
#     y.append(countBitstringArea(grover(6,imgdata)))
    axes[i//8*2,i%8].imshow(imgdata, cmap='gray')
    axes[i//8*2,i%8].set_xticks([])
    axes[i//8*2,i%8].set_yticks([])
    
    g = imigize1(grover(6,imgdata))
    large = []
    for row in g:
        large.append(np.repeat(row, 2))
    large = np.repeat(large, 2, axis=0)
    image=np.asarray(large, dtype=np.float32)
    plt.imsave(str(i)+".jpg", image)
    axes[i//8*2+1,i%8].imshow(image, cmap='gray')
    axes[i//8*2+1,i%8].set_xticks([])
    axes[i//8*2+1,i%8].set_yticks([])
    print(str(i) + 'th iteration:',time.ctime())
```
